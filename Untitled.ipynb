```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(image, target_size=(224, 224)):
    # Görseli boyutlandır
    image = image.resize(target_size)
    
    # Görseli normalize et
    image = np.array(image) / 255.0
    
    return image

# Veri setinin ana klasör yolu
folder = r"C:\Users\Admin\Desktop\proje\archive (3)"

# 'train' ve 'test' klasörlerinin altındaki 'real' ve 'fake' klasörlerini kontrol et
subfolders = ['train/real', 'train/fake', 'test/real', 'test/fake']

# Görsel yolları ve etiketleri depolayacağımız listeler
image_paths = []  # Görsel yolları
labels = []  # Etiketler (0: fake, 1: real)

# Klasörleri gezerek dosya yollarını öğrenme
for subfolder in subfolders:
    folder_path = os.path.join(folder, subfolder)  # Alt klasörün tam yolu
    jpg_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg')]  # JPG dosyalarını seçme
    
    for file in jpg_files:
        # Her bir görselin tam yolunu oluşturma
        image_path = os.path.join(folder_path, file)
        image_paths.append(image_path)
        
        # Etiket belirleme: 'real' dosyaları için 1, 'fake' dosyaları için 0
        label = 1 if 'real' in subfolder else 0
        labels.append(label)

# İlk 5 görseli ön işleme yaparak gösterme
for i in range(5):
    # Görseli aç
    image = Image.open(image_paths[i])
    
    # Görseli işleme
    processed_image = preprocess_image(image)
    
    # Görseli matplotlib ile gösterme
    plt.subplot(1, 5, i+1)  # 1 satır, 5 sütun
    plt.imshow(processed_image)
    plt.axis('off')  # Eksenleri gizle
    plt.title(f"Label: {labels[i]}")  # Etiketi başlık olarak ekle

plt.show()  # Görselleri göster
```

```python
import os
import cv2
import numpy as np


data_dir = r"C:\Users\Admin\Desktop\proje\archive (3)"
categories = ["real", "fake"]  


def load_data(folder_name):
    data = []
    folder_path = os.path.join(data_dir, folder_name)  # train veya test dizini
    for category in categories:
        category_path = os.path.join(folder_path, category)  # real veya fake dizinine gider
        class_label = categories.index(category)  # 0 veya 1
        for img_name in os.listdir(category_path):
            try:
                img_path = os.path.join(category_path, img_name)
                img = cv2.imread(img_path, cv2.IMREAD_COLOR)
                img = cv2.resize(img, (128, 128))  # Tüm görüntüleri 128x128 boyutunda yeniden boyutlandırdım
                data.append((img, class_label))
            except Exception as e:
                print(f"Hata oluştu: {e}, dosya: {img_path}")
    return data

# Train ve test verilerini yükledim
train_data = load_data("train")
test_data = load_data("test")

# Veriyi ve etiketleri ayrıştırdım
X_train, y_train = zip(*train_data)
X_test, y_test = zip(*test_data)

# normalizasyon yaptım
X_train = np.array(X_train) / 255.0
X_test = np.array(X_test) / 255.0
y_train = np.array(y_train)
y_test = np.array(y_test)

print("Veriler başarıyla yüklendi!")
print(f"X_train boyutu: {X_train.shape}")
print(f"X_test boyutu: {X_test.shape}")
print(f"y_train boyutu: {y_train.shape}")
print(f"y_test boyutu: {y_test.shape}")
```

```python
import os
import cv2
import numpy as np
import random
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from skimage.feature import hog, local_binary_pattern
import pandas as pd

# LBP parametreleri
LBP_RADIUS = 3
LBP_POINTS = 8 * LBP_RADIUS

# HOG özellik çıkarımı
def extract_hog_features(image):
    if image.dtype != np.uint8:
        image = np.uint8(image)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    features, _ = hog(gray_image, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)
    return features

# Renk Histogramı özellik çıkarımı
def extract_color_histogram(image, bins=(8, 8, 8)):
    hist = cv2.calcHist([image], [0, 1, 2], None, bins, [0, 256, 0, 256, 0, 256])
    cv2.normalize(hist, hist)
    return hist.flatten()

# LBP özellik çıkarımı
def extract_lbp_features(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lbp = local_binary_pattern(gray_image, LBP_POINTS, LBP_RADIUS, method="uniform")
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, LBP_POINTS + 3), range=(0, LBP_POINTS + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-6)
    return hist

# HOG Renk Histogramı LBP özelliklerini birleştirdim
def extract_combined_features(image):
    hog_features = extract_hog_features(image)
    color_hist_features = extract_color_histogram(image)
    lbp_features = extract_lbp_features(image)
    return np.hstack((hog_features, color_hist_features, lbp_features))

# Görsellerden örnekleme yaptım ve özellikleri çıkardım
def sample_images(folder_path, num_samples):
    sampled_features = []
    sampled_labels = []
    for label, class_folder in enumerate(["FAKE", "REAL"]):
        class_folder_path = os.path.join(folder_path, class_folder)
        all_images = os.listdir(class_folder_path)
        sampled_images = random.sample(all_images, num_samples)
        for file_name in sampled_images:
            file_path = os.path.join(class_folder_path, file_name)
            if os.path.isfile(file_path):
                image = cv2.imread(file_path)
                if image is not None:
                    sampled_features.append(extract_combined_features(image))
                    sampled_labels.append(label)
    return np.array(sampled_features), np.array(sampled_labels)

# Karışıklık matrisini çizdim
def plot_confusion_matrix(cm, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(2)
    plt.xticks(tick_marks, ['FAKE', 'REAL'])
    plt.yticks(tick_marks, ['FAKE', 'REAL'])
    plt.ylabel('Gerçek Etiket')
    plt.xlabel('Tahmin Etiketi')
    plt.tight_layout()

train_folder = r'C:\Users\Admin\Desktop\proje\archive (3)\train'
test_folder = r'C:\Users\Admin\Desktop\proje\archive (3)\test'

# Eğitim ve test veri setlerini hazırladım
train_features, train_labels = sample_images(train_folder, 2000)
test_features, test_labels = sample_images(test_folder, 1000)

# Özellikleri ölçekleledim
scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)

X_train, X_test, y_train, y_test = train_test_split(train_features, train_labels, test_size=0.2, random_state=42)

# Model tanımları
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
knn_model = KNeighborsClassifier(n_neighbors=5)
svm_model = SVC(kernel='linear', random_state=42, probability=True)

# Model eğitme ve değerlendirme
models = {
    "Random Forest": rf_model,
    "KNN": knn_model,
    "SVM": svm_model
}

performance_data = []

for model_name, model in models.items():
    print(f"\n{model_name} Modeli Eğitiliyor...")
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Model Sonuçları:")
    print(classification_report(y_test, predictions))
    cm = confusion_matrix(y_test, predictions)
    plt.figure(figsize=(10, 7))
    plot_confusion_matrix(cm, title=f'{model_name} Confusion Matrix')
    plt.show()
    performance_data.append({"Model": model_name, "Accuracy": accuracy})

# Performans karşılaştırma tablom
performance_df = pd.DataFrame(performance_data)

print("\nModel Performans Karşılaştırması:")
print(performance_df)

# Performans tablosunu görselleştirme
plt.figure(figsize=(10, 7))
plt.bar(
    performance_df['Model'],
    performance_df['Accuracy'],
    color=['#FF5733', '#5DADE2', '#58D68D']  
)
plt.xlabel('Model')
plt.ylabel('Doğruluk')
plt.title('Model Performans Karşılaştırması')
plt.ylim(0, 1)
plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt


train_folder = r'C:\Users\Admin\Desktop\proje\archive (3)\train'
test_folder = r'C:\Users\Admin\Desktop\proje\archive (3)\test'

# Parametreler
IMG_HEIGHT, IMG_WIDTH = 128, 128  
BATCH_SIZE = 32
EPOCHS = 15

# Veri ön işleme
train_datagen = ImageDataGenerator(rescale=1.0/255.0,
                                   rotation_range=20,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1.0/255.0)

train_generator = train_datagen.flow_from_directory(
    train_folder,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_folder,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

# CNN Modeli
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    MaxPooling2D((2, 2)),
    
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Model derleme
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Model özeti
model.summary()

# Modeli eğitme
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=test_generator
)

# Performans görselleştirme
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.title('Model Doğrulukları')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.title('Model Kayıpları')
plt.legend()
plt.show()

# Tahminler ve raporlama
y_pred = model.predict(test_generator)
y_pred_classes = np.where(y_pred > 0.5, 1, 0)

print("\nSınıflandırma Raporu:")
print(classification_report(test_generator.classes, y_pred_classes))

cm = confusion_matrix(test_generator.classes, y_pred_classes)
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Karışıklık Matrisi')
plt.colorbar()
tick_marks = np.arange(2)
plt.xticks(tick_marks, ['FAKE', 'REAL'])
plt.yticks(tick_marks, ['FAKE', 'REAL'])
plt.ylabel('Gerçek Etiket')
plt.xlabel('Tahmin Etiket')
plt.show()
```
